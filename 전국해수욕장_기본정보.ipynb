```python
url = "http://apis.data.go.kr/1192000/service/OceansBeachInfoService1/getOceansBeachInfo1?pageNo=1&numOfRows=100&resultType=xml&SIDO_NM={sido}&ServiceKey={key}"
```

```python
sido_list = ["부산","인천","울산","강원","충남","전북","전남","경북","경남","제주"]
beach_df = pd.DataFrame()

for y in range(len(sido_list)) :
    
    # url에 파라메터 값 정의
    api_url = url.format(sido = sido_list[y], key = apikey)

    # url을 이용해서 요청
    rs_data = requests.get(api_url)

    # 1. 텍스트로 변환
    rs_text = rs_data.text

    # 2. xml 문서 포맷 정의
    rs_xml = BeautifulSoup(rs_text,"lxml")

    # 3. 값 추출
#     for i in range(len(rs_xml.find_all("stanm"))):
#         print("일련번호 : ", rs_xml.find_all("num")[i].text)
#         print("시도명 : ", rs_xml.find_all("sidonm")[i].text)
#         print("구군명 : ", rs_xml.find_all("gugunnm")[i].text)
#         print("정점명 : ", rs_xml.find_all("stanm")[i].text)
#         print("해변폭 : ", rs_xml.find_all("beachwid")[i].text)
#         print("해변총연장 : ", rs_xml.find_all("beachlen")[i].text)
#         print("특징 : ", rs_xml.find_all("beachknd")[i].text)
#         print("관련사이트 : ", rs_xml.find_all("linkaddr")[i].text)
#         print("관련사이트 이름 : ", rs_xml.find_all("linknm")[i].text)
#         print("해수욕장 비상연락처 : ", rs_xml.find_all("linktel")[i].text)
#         print("위도 : ", rs_xml.find_all("lat")[i].text)
#         print("경도 : ", rs_xml.find_all("lon")[i].text)
#         print("--------------------------------------")
    for i in range(0, len(rs_xml.find_all("stanm")),1):
        a =len(beach_df)
        beach_df.loc[a,"일련번호"] = rs_xml.find_all("num")[i].text
        beach_df.loc[a,"시도명"] = rs_xml.find_all("sidonm")[i].text
        beach_df.loc[a,"구군명"] = rs_xml.find_all("gugunnm")[i].text
        beach_df.loc[a,"정점명"] = rs_xml.find_all("stanm")[i].text
        beach_df.loc[a,"해변폭"] = rs_xml.find_all("beachwid")[i].text
        beach_df.loc[a,"해변총연장"] = rs_xml.find_all("beachlen")[i].text
        beach_df.loc[a,"특징"] = rs_xml.find_all("beachknd")[i].text
        beach_df.loc[a,"관련사이트"] = rs_xml.find_all("linkaddr")[i].text
        beach_df.loc[a,"사이트이름"] = rs_xml.find_all("linknm")[i].text
        beach_df.loc[a,"비상연락처"] = rs_xml.find_all("linktel")[i].text
        beach_df.loc[a,"위도"] = rs_xml.find_all("lat")[i].text
        beach_df.loc[a,"경도"] = rs_xml.find_all("lon")[i].text
beach_df
```

```python
df = beach_df.copy()
# 결측치 처리 
df.replace("", 0,inplace=True)
df["비상연락처"].replace(0, "알수없음",inplace=True)
df["특징"].replace(0, "알수없음",inplace=True)
df
```

```python
## 형태 변환
df["일련번호"] = df["일련번호"].astype(int)
df["해변폭"] = df["해변폭"].astype(int)
df["해변총연장"] = df["해변총연장"].astype(int)
df["위도"] = df["위도"].astype(float)
df["경도"] = df["경도"].astype(float)
```
