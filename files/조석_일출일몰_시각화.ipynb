# 일출시각 평균치 구해서 시각화하기

```python
# 일출시각에 결측치가 있는 데이터를 제외
sun_temp = sun_merge_df_no_du[(sun_merge_df_no_du["일출시각"]=="00:00") ==False].reset_index(drop=True)
sun_temp
```

### 결측치 제외한 일출시각의 평균시각 구하기

```python
# 시간 데이터로 평균내는 방법은 datetime 을 이용해서 직접 시간, 분, 초를 초로 전부 변환하고 
# 이것을 평균내서, 다시 시간으로 반환하는 방법밖에.. 

# sun_temp["일출시각"].mean()

sun_rise_time_list = []

for i in range(len(sun_temp["해수욕장이름"])) : 
    
    #일출시각 가져오기
    wt = (sun_temp["일출시각"][i])
    
    #일출시각에서 시간, 분을 추출하여 datetime 형식으로 변환
    sun_rise_time = datetime.time(int(wt[:2]),int(wt[3:]),0)
    
    #datetime형식의 일출시각을 빈 리스트에 추가 
    sun_rise_time_list.append(sun_rise_time)
    

#datetime형식이 들어있는 리스트에서 시간, 분, 초를 '초'로 반환하여 모두 더하기
sun_rise_time_total = sum(t.hour * 3600 + t.minute * 60 + t.second for t in sun_rise_time_list)

#더한 값을 갯수로 나누어 평균 구하기 
sun_rise_average =sun_rise_time_total / len(sun_rise_time_list)

# 평균시각(초 형태)를 datetime 형식으로 변환
kk = datetime.time(int(sun_rise_average/3600), int((sun_rise_average % 3600) / 60), int(sun_rise_average % 60))

kk

# print(len(sun_rise_time_list))
```
