# 하늘상태 시각화

```python
sky_map = folium.Map(

    location = [36.684273, 128.068635],
    tiles = "openstreetmap",
    zoom_start = 6.5 )

# 지도에 Mark 찍기
for i in range(len(merge_df["해수욕장이름"])) : 
    
    lat = merge_df.loc[i,"위도_x"]
    lng = merge_df.loc[i,"경도_x"]
    wt = merge_df["하늘상태"][i]
    sky_text = "[{beach}] <br/> 하늘상태 : {sky} ".format(sky = merge_df.loc[i,"하늘상태"], beach =merge_df.loc[i,"해수욕장이름"] )
    

    if wt  == "맑음":
        m_icon = folium.Icon(icon = 'sun',prefix = 'fa',icon_size=(1, 1), icon_color = "yellow",shadow_size=(0, 0),icon_anchor=(4, 30))
    elif wt  == "구름많음" :
        m_icon = folium.Icon(icon = "cloud",icon_size=(7, 7), icon_color = "white",shadow_size=(0, 0),icon_anchor=(4, 30))
    elif wt  == "흐림"  :
        m_icon = folium.Icon(icon = "cloud", icon_size=(7, 7), icon_color = "black",shadow_size=(0, 0),icon_anchor=(4, 30))
    else :
        m_icon = folium.Icon(icon_size=(7, 7), opacity=0)
        

    folium.Marker(
    icon = m_icon,
    location = [lat,lng],
    tooltip = sky_text
    ).add_to(sky_map)
    
sky_map
```
